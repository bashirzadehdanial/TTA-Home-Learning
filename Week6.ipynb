```python
points=alt.Chart(cars).mark_point().encode(
    x='Miles_per_Gallon:Q',
    y='Acceleration:Q',
    color='Acceleration').add_selection(brush)
points
```

```python
bar=alt.Chart(cars).mark_bar().encode(
    x='Cylinders:Q',
    y='Weight_in_lbs:Q',
    color='Weight_in_lbs').add_selection(brush)
bar
```

```python
points= alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color='Horsepower').add_selection( brush)
bar = alt.Chart(cars).mark_bar().encode(
    y='Origin:N',
    color='Origin:N',
    x='count(Origin):N',).transform_filter(brush)
points & bar
```
